```python
gimg = cv2.imread('/content/gdrive/My Drive/IPColab/images/images/abcdef.bmp')
print(gimg.shape)
gimg = cv2.cvtColor(gimg, cv2.COLOR_BGR2GRAY)
plt.imshow(gimg, cmap='gray')
#cv2_imshow(gimg)
```

```python
numLabels, labels = cv2.connectedComponents(gimg)
print(numLabels)
```

```python
print(labels.shape)
print(labels)
```

```python
label_H = np.uint8(labels * 179 /np.max(labels))
cv2_imshow(label_H)
```

```python
bank = 255*np.ones_like(label_H)
labelimg = cv2.merge([label_H, bank, bank])
labelimg = cv2.cvtColor(labelimg, cv2.COLOR_HSV2RGB)
cv2_imshow(labelimg)
```
