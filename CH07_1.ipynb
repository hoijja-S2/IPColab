```python
from google.colab.patches import cv2_imshow
from google.colab import drive
import numpy as np #이미지는 기본적으로 2차원 배열이기 때문
import cv2
import matplotlib.pyplot as plt
drive.mount('/content/gdrive')
```

```python
gimg = cv2.imread('/content/gdrive/My Drive/IPColab/images/images/lena256.jpg')
print(gimg.shape)
gimg = cv2.cvtColor(gimg, cv2.COLOR_BGR2GRAY)
plt.imshow(gimg, cmap='gray')
#cv2_imshow(gimg)
```

```python
outimg = np.zeros(gimg.shape, dtype=np.uint8)
cv2_imshow(outimg)
```

```python
scaling = 1.5

for h in range(gimg.shape[0]):
  for w in range(gimg.shape[1]):
    y = int(h * scaling)
    x = int(w * scaling)
    if y < gimg.shape[0] and x < gimg.shape[1]:
      outimg[y, x] = gimg[h, w]

cv2_imshow(outimg)
```

```python
scaling = 3

for y in range(gimg.shape[0]):
  for x in range(gimg.shape[1]):
    h = int(y / scaling)
    w = int(x / scaling)
    if h < gimg.shape[0] and w < gimg.shape[1]:
      outimg[y, x] = gimg[h, w]

cv2_imshow(outimg)
```

```python
height, width = gimg.shape
print(height, width)
```

```python
M = cv2.getRotationMatrix2D((width/2.0, height/2.0), 45, 1)
print(M)
```

```python
outimge = cv2.warpAffine(gimg, M, (width, height))
cv2_imshow(outimge)
```
